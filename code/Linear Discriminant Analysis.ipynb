```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh
```

```python
# toy example: 3 classes (1, 2, and 3), 300 points in R2
X = np.zeros((2,300))
y = np.zeros(300)

# class 1 points:
X[:,:100] = np.random.randn(2,100) + np.array([[5],[5]])
y[:100] = 1

# class 2 points:
X[:,100:200] = np.random.randn(2,100) + np.array([[0],[10]])
y[100:200] = 2

# class 3 points:
X[:,200:] = np.random.randn(2,100) + np.array([[0],[-3]])
y[200:] = 3

# plot data:
plt.scatter(X[0,y==1],X[1,y==1],c='k')
plt.scatter(X[0,y==2],X[1,y==2],c='r')
plt.scatter(X[0,y==3],X[1,y==3],c='b')
plt.show()
```
