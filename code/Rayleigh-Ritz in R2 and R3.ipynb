```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm # matplotlib color maps
from scipy.linalg import eigh # computes eigenvalues/eigenvectors of symmetric matrices (or Hermitian)
```

```python
A = np.array([[5,2],[2,-1]])
A
```

```python
# compute eigenvalues and eigenvectors of A
D,Q = eigh(A)
q1, q2 = Q
```

```python
# generate m unit vectors
m = 1000
thetas = np.linspace(0,2*np.pi,m)
uvecs = np.array([np.array([np.cos(theta),np.sin(theta)]) for theta in thetas]).T
# evaluate the quadratic form at the unit vectors
f = np.sum(uvecs*A.dot(uvecs),axis=0)
```

```python
# plot the quadratic form on the unit circle
plt.scatter(uvecs[0], uvecs[1], c=f, cmap='plasma')
plt.colorbar()
# plot the eigenvectors
plt.arrow(0,0,q1[0],q1[1])
plt.arrow(0,0,q2[0],q2[1])
plt.show()
```
